```python
import numpy as np
import matplotlib.pyplot as plt
import waipy
from waipy import wavelet_analysis

def create_wave_data(length=1000, freq=1):
    # Generate a sample time array
    time = np.arange(0, length)
    # Create a sine wave
    data = np.sin(2 * np.pi * freq * time / length)
    return time, data

# Generate data
time, data = create_wave_data(length=1000, freq=5)

# Perform Wavelet Transform
result = wavelet_analysis(data, time, dt=1, pad=1, dj=0.25)

# Plotting
plt.figure(figsize=(12, 6))
plt.subplot(211)
plt.title('Original Signal')
plt.plot(time, data, color='blue')
plt.xlim([0, len(data)])

plt.subplot(212)
plt.title('Wavelet Power Spectrum')
waipy.cwt_graphic(result, time=time, title='Wavelet Transform of sine wave')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import waipy


N = 2000
t = np.linspace(0, 200, N)
f = np.sin(2 * np.pi * t / 5) + np.sin(2 * np.pi * t / 3) + np.random.normal(0, 0.1, N)

s0 = 1  # Smallest scale (e.g., daily cycles)
s1 = 10  # Largest scale (e.g., longer-term patterns)
ds = 0.1  # Scale step
nvoice = 6  # Number of octaves per scale
alpha = 6  # Alpha parameter for Morlet wavelet
mother = 'Morlet'  # Wavelet type

# Perform wavelet analysis
result = waipy.cwt(f, s0, s1, ds, nvoice, (s1-s0)/ds, alpha, 6, mother=mother, name="Morlet")

label = 'Total 21-25 cycle'
time = t
plt.figure(figsize=(10, 6))
waipy.wavelet_plot(label, time, f, ds, result, plot_power=False, contour_levels=20, yscale='linear', ylabel='Period')
# plt.savefig("/content/drive/MyDrive/Solar_index/output/Wavelet_21-25_cycle_Total.png")
plt.show()
```
